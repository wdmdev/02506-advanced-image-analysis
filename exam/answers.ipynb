```python
filename = 'EXAM_DATA_2021/frame.png'
img = io.imread(filename).astype(float)/255
plt.imshow(img, cmap='gray')
```

```python
center = (img.shape[0]/2, img.shape[1]/2)
center
```

```python
mask = np.zeros(img.shape, dtype=bool)

for i in range(int(center[0]-40), int(center[0]+40)):
    for j in range(int(center[1]-40), int(center[1]+40)):
        mask[i,j] = 1
```

```python
m_in = np.mean(img[mask])
m_out = np.mean(img[~mask])

p = [img.shape[0]/2+39.5, img.shape[1]/2-40.5]
I_p = img[int(p[0]),int(p[1])]
f_ext = (m_in - m_out) * (2*I_p - m_in - m_out)
```
